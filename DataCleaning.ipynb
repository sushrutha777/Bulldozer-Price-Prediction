```python
# Define a mapping dictionary
product_size_mapping = {
    "Large / Medium": "Large",
    "Large": "Large",
    "Mini": "Small",
    "Small": "Small",
    "Medium": "Medium",
    "Compact": "Compact",
    "Unknown": "Unknown"
}

# Apply mapping
df["ProductSize"] = df["ProductSize"].map(product_size_mapping)

# Check new value counts
print(df["ProductSize"].value_counts())
```

```python
# Select only object (categorical) columns
cat_cols = df.select_dtypes(include=["object"]).columns

# Apply value_counts() to each categorical column
for col in cat_cols:
    print(f"\n🔹 Column: {col}")
    print(df[col].value_counts(), "\n" + "-"*50)
```

```python
df.drop(columns=["Forks", "Ride_Control", "Transmission", "Hydraulics", "Coupler"], inplace=True)
```

```python
df.shape
```

```python
df.fiBaseModel.nunique()
```

```python
#The code replaces fiBaseModel values that appear less than 100 times with "Other", 
#reducing high cardinality for better model performance.
base_model_counts = df["fiBaseModel"].value_counts()
rare_models = base_model_counts[base_model_counts < 100].index
df["fiBaseModel"] = df["fiBaseModel"].apply(lambda x: "Other" if x in rare_models else x)
```

```python
df['MachineHoursCurrentMeter'] = df['MachineHoursCurrentMeter'].astype(int)
df['auctioneerID'] = df['auctioneerID'].astype(int)
```

```python
df.info()
```

## Encoding

```python
from sklearn.preprocessing import LabelEncoder

# Label Encoding (For ordinal categories like ProductSize)
label_encoder = LabelEncoder()
df["ProductSize"] = label_encoder.fit_transform(df["ProductSize"])

# One-Hot Encoding (For nominal categories like state, ProductGroup, ProductGroupDesc, Enclosure)
df = pd.get_dummies(df, columns=["state", "ProductGroup", "ProductGroupDesc", "Enclosure"], drop_first=True)

# Target Encoding (For high-cardinality columns: fiBaseModel, fiProductClassDesc)
for col in ["fiBaseModel", "fiProductClassDesc"]:
    df[col] = df.groupby(col)["SalePrice"].transform("mean")
```

```python
# Conert True -> 1 , false -> 0
bool_cols = df.select_dtypes(include="bool").columns  # Select boolean columns
df[bool_cols] = df[bool_cols].astype(int)  # Convert only bool columns
```

```python
df['fiBaseModel'] = df['fiBaseModel'].astype(int)
df['fiProductClassDesc'] = df['fiProductClassDesc'].astype(int)
```
